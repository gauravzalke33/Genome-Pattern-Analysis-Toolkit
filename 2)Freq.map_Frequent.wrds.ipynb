# FREQUNCY MAPPING

this function will return us all the possible k-mers from the provided gene or text:

here we will be running two for loop :

1) the first loop is to get all the possibe k-mers and since we will be using dictionary for this function after running our first loop it will assign the value of zero to all possible k-mers in the text
2) now the second loop will be responsible for the increment of the number of the values one by one , on the occurence of the same k-mer

we will bw running a window here of the provided k here inorder to get all the possible k-mers in the given genme or text 

```python
def FrequencyMap(Text, k):
    freq = {}
    n = len(Text)
    for i in range(n-k+1):
        Pattern = Text[i:i+k]
        freq[Pattern] = 0
    return freq

FrequencyMap("AAGCGTAACGCTCAGCTA",3)

# since this is our first loop it will only return all the k-mers without mapping it!!!!!
```

now lets do this with the second loop too , to get the incremented values 

```python
def FrequencyMap(genome,k):
    freq = {}
    for i in range(len(genome)-k+1):
        pattern = genome[i:i+k]
        freq[pattern] = 0
    for i in range(len(genome)-k+1):
        pattern = genome[i:i+k]
        freq[pattern] += 1
    return freq
FrequencyMap("AAGCGTAACGCTCAGCTA",3)
```

now its perfect cause we added one more for loop to increase freq[pattern] by one time each 

first loop to get all the possible k-mers

and second loop to count the occurence one bt one 

```python
# frequency map with single loop
def map(genome,k):
    freq = {}
    for i in range(len(genome)-k+1):
        pattern = genome[i:i+k]
        if pattern not in freq:
            freq[pattern] = 1
        else:
            freq[pattern] += 1
    return freq
map("AAGCGTAACGCTCAGCTA",3)
```

# FREQUENT WORDS

```python
def FrequentWords(Text, k):
    words = []
    freq = FrequencyMap(Text, k)
    m = max(freq.values())
    for key in freq:
        # add each key to words whose corresponding frequency value is equal to m
        if freq[key] == m :
            words.append(key)
    return words
```
