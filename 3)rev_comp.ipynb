First, we will tackle Reverse().  We can range over each character char of a string Pattern (from beginning to end) using the notation

for char in Pattern:

Code Challenge (2 Points): Write a function Reverse() that takes a string Pattern and returns a string formed by reversing all the letters of Pattern.  Recall that for strings x, y, and z, the notation z=x+y concatenates x and y together and stores the result in the variable z.

 Hint: We can begin with an empty string rev and then successively place each character of Pattern at the beginning of this string.  For example, if we were taking the reverse of ACGT, then rev would grow as follows.

A;
CA;
GCA;
TGCA

Sample Input:
AAAACCCGGT

Sample Output:
TGGCCCAAAA

```python
# Input:  A string Pattern
# Output: The reverse of Pattern
def Reverse(Pattern) :
    rev = ''
    for char in Pattern:
        rev = char + rev
    return rev
Reverse("AAAAGGGCCT")
```

```python
# another way: use the array's slice methods to reverse it

def Reverse(Pattern):
    return Pattern[::-1]

Reverse("AAAAGGGCCT")
```

Initialization: rev = "" creates an empty string called rev. This string will store the reversed version of the input Pattern.

Iteration: The for char in Pattern: loop iterates through each character (char) in the input string Pattern, from the beginning to the end.

Concatenation: Inside the loop, the line rev = char + rev is the key to reversing the string. It does the following:

It takes the current character char.

It concatenates char to the beginning of the existing rev string.

The result is then assigned back to rev.

Example Breakdown (for "ACGT"):

Initially, rev is "".

1st iteration: char is 'A'. rev becomes "A" + "" = "A".

2nd iteration: char is 'C'. rev becomes "C" + "A" = "CA".

3rd iteration: char is 'G'. rev becomes "G" + "CA" = "GCA".

4th iteration: char is 'T'. rev becomes "T" + "GCA" = "TGCA".

Return: After the loop has processed all characters, the function returns the rev string, which now contains the reversed version of the input Pattern

# Pattern Matching Problem: Find all occurrences of a pattern in a string.

Input: Strings Pattern and Genome.

 Output: All starting positions in Genome where Pattern appears as a substring.

Sample Input:
ATAT;;;;;;
GATATATGCATATACTT

Sample Output:
1 3 9


```python
# fill in your PatternMatching() function along with any subroutines that you need.
def PatternMatching(Pattern, Genome):
    positions = [] # output variable
    for i in range(len(Genome)-len(Pattern)+1):
        if Genome[i:i+len(Pattern)] == Pattern :
            positions.append(i)
    return positions

PatternMatching("ATAT","GATATATGCATATACTT")
```
