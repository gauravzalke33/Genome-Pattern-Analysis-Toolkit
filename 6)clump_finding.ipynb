- Text : the genome provided
- k    : the length of pattern or k-mer
- L    : the length of the window
- t    : the minimum number if times it should occure in a window of lenght L 

in clump finding lets assume the the len of text is 100 and the given lenght of the window is 25 ,
so its not necessary that the clump should occure in the every window of the Text , 
even if it occured for the t times in one window and never showed in any other window or showed up for less than t times in the window it can be still called as a clump

## STEPS

- we will start with making an empty list
- then we will run a window of given lenght L in the Genome
- the we will assign a variable (`window`)
- then we will make a `freq` variable for `freqmap` 
- then for keys or `k-mers` in that map we will check if the values for those keys are more or equal to `t` or not 
- and also that those keys are not already pressent in the empty list `pattern` , this can we skiped if we use empty set instead of an empty list to avoid the possibility of multiple occurence
- then we will append the key in our empty list 
- and finally return

````python
def ClumpFinding(Genome,k,L,t):
    clumps = []
    for i in range(len(Genome)-L+1):
        window = Genome[i:i+L]
        freq = frequencymap(window,k)
        for key in freq:
            if freq[key] >= t and key not in clumps:
                clumps.append(key)
    return clumps
def frequencymap(genome,k):
    freq = {}
    for i in range(len(genome)-k+1):
        pattern = genome[i:i+k]
        freq[pattern] = 0
    for i in range(len(genome)-k+1):
        pattern = genome[i:i+k]
        freq[pattern] += 1
    return freq

print(*ClumpFinding("CTAGGTCAGCTACCTAGATCTACACAGGACTCACATCAGGACTCACATATGCCATGCGGTAAAAGTGTTCCCAGGCACAGGACTCACACAGGACTCACGCAGGACTCACACTTCTACAAGGCGCAAGTGTCCGTTTGGGCTTATAGTTTGGGCTTTCTGAAATGAAGTTTGGGCTTACGGTTTGGGCTTCTTTTTGGGCTTCCAGTTTGGGCTTGACTCTCTTTCGGGCGCATGAATCGCAATTTAAACTTCTTGAATGAACAGATGGTGATAAATCATGCCCTAGTCCATGGTAGGAAAACTTGCAAAGCGAGTTGTGCGAACGTGACATTGCCAACGGAGGGCACGTGGGAGAGCTGGGTAGGAGAATCCCTCTCTGTCAAGTTCCTCCGCGGGGCGCTGTTCCATTATACGGACAAAGCATTAGTTGAGTGAGGAAATAAGGCATTAATTAGTTGAGATTAGTTGAGACAAATTAGTTGAGACCCACAGTGTGAACAACGGTCTTATCACAGCTGCCGACCACGCGGACCTTCTCATGTAAAGCCATCGGTGTAGGGCTAACGGGTGGTCTTGGAGGTAGTTCTAATGGACCGCGAGCACAACCGGTTCAGAGAAAATCTGAGGGCCGTGTCAGCTATTCTTATGCAAGGGTAGTAATGTTTTCATATCTGCCTGATTTTCAATTGTCAAACGGATTCGTCGGGCCAGAAGAAGCAAAGCCATAACAAGCCTTATGTCACTTGTCACTTTGAAGAAACCTTGTGTCACTTTGTGTTTCAGATAGCAGACCCACGTGTGTCACTTTGTCACTTTGTGTCTGTCACTTTGTTTAAAATCTCGACGCCGATTAAATCTCGACCTAAATCTCGACAATGCTAGATCAGTCTCGCAAACATAATATTCGAAAATCTCAAATCTCGACTCGTATGTTCTACAAAGGCTAGGGCAGGCTCTTTGTTCTTTGTGTTTGAGGACCTTGTCTGCGGTCGATGGCTCGGCTCTGGCTCTTTGTTGGACGGATGGACGGAGGCTCTTTGTCCGTTGGACGGATAATAATTCAGTTGCGTGGACGGATATGATAGCTGATCGATGAGGCCCCCAAGGCCAGGCCCCCAGCACAGGCCCCCAAGGCCCCCAGACACAAACTATCATCAGGCAGGCCCCCAGAAGGCCCCCAGTATCACCTTCACTTCTTAGGGTGCACCTCGGTGCGACGATTAACGGTGCGGTCGGTGCGACGAAGACCGGTGCGACCGGTGCGACGGACGAAATATCCTGGAGTTGACGGTGCGACGAGCTTTCGGTCATTGGCGCTGTTTCATCTGCAGTGGCGCCGGTACCCCATTCTCGTAAGATGGAGTCTGTGGTTGATACATATGAGCCTACTACGCACTGAGTATACTCGCTATCGTGTGGTTCTCGCCCTCGGTGTAGTCGAGAAATCCAGCTGAAATCGTGACAAGCTCAGCGGACTCCGGCTAAAGAGTAAAATTGCTGGGTTGTTGCGCTGCTCTAACAGACTCGGGTGTGAACGAGAGGATCTCCACATCGCCTACCGCTGACCAGGAAATATACCTAGGCATATCCCCTGGGAGGCTCTATTTTAAGCGCGGGCGTCAGACCTGTGTCGTAGGGCCTCCCGATCTGTCATTACATTACTGCGGTTAATATGGAACGGAGAGTCACGACTCTTGGCACGTAAGGCCAGGCCTATGGCTGGAACGTAGCAATAAAGTTAACTTAGCCGGAATCTAGCCCCTATTGGACAAAGGATCGTTTAACATATCGGACCGCCCCCTAGTCCATCTCATGCCCATAGTGTGGTCTTAACCGTACCATCTCATGCCCCATCTCATGCATCTCATGCATGCCCTCATGCTTGTGACCGAGCGGGATAAAGGCATATGTGGGTGTTTCGTCAGCTACCCATAGTGTTCACAGGTAAGGGCAAGTGCCTTAGCTGTGGAGCTCGAGTGTAATCTAACATATTGTTTAAC",10,100,4),sep = " ")
````
